Let's explore how numba generates multiple LLVM modules 

```python
import numba
import os
```

```python
os.environ['NUMBA_DUMP_OPTIMIZED'] = '1'
```

```python
@numba.njit
def hi():
    print("hi")
```

```python
@numba.jit
def there():
    print("there")
    hi()
```

```python
there()
```

We notice these modules are actually combined in the `there` module. `__main__9there` contains both print statements
